# 건강검진정보를 기반의 국민의 음주 문화 분석

# 성별, 나이별 음주인원 수 히트맵

```python
df_age =  pd.pivot_table(df, index = ['연령대 코드(5세단위)'],columns=['성별코드'] ,values = '음주여부', aggfunc = 'count')
df_age
```

```python
# 히트맵

plt.figure(figsize=(15,12))
sns.heatmap(df_age, cmap='YlGn',annot=True, fmt='d',annot_kws={"size":15})
plt.xlabel('성별', fontsize=16)
plt.ylabel('나이 (단위*5)', fontsize=16)
plt.title('성별, 나이별 음주자 수', fontsize=20)
```

위 그래프는 나이대/ 성별 음주인원 수 히트맵이다.

나이 코드는 5세 단위이기 때문에 나이코드 * 5 를 하면 나이를 알 수 있다.

65세 이전까지는 남성이 더 많지만 65세 이후로는 비슷하지만 여성의 수가 더 많아지는 것을 확인할 수 있다.

남성의 경우 65세까지는 7만명대 이지만 65세 이후 4만명으로 감소하는 것을 볼 수 있다.

여성의 경우 55세 전까지 6만5천명대를 유지하지만 55 ~ 65 세 까지는 남성과 비슷할 정도로 늘어나는 것을 확인할 수 있다.

# 음주 여부에 따른 체중 히스토그램

```python
df_waist = df[(df['허리둘레']<200) & (df['총 콜레스테롤']<600)]
```

```python
plt.figure(figsize=(15,12))
sns.histplot(data = df_waist, x = df_waist['체중(5Kg 단위)'],hue = '음주여부', palette=['#ACFA58', '#FA5858'],bins=20)
plt.xlabel('체중 5kg 단위', fontsize=16)
plt.ylabel('인구 수', fontsize=16);
plt.title('음주 여부 비교그래프', fontsize=20)
```

위의 히스토그램을 보았을 때 음주를 하는 사람들이 전반적으로 고체중인 경우가 많은 것을 확인할 수 있었습니다.

대략 60kg 이상인 사람들 이상으로는 전부 음주를 하는 사람들이 많은 것을 확인할 수 있엇고

체중이 많아질수록 음주인원과 비 음주인원의 차이도 커지는 것을 확인할 수 있었습니다.



# 음주 여부에 따른 체중별 콜레스테롤 라인 그래프

```python
# line
plt.figure(figsize=(15,12))
sns.lineplot(data = df, x=df['체중(5Kg 단위)'],y = df['총 콜레스테롤'],palette=['#ACFA58', '#FA5858'],hue = '음주여부',ci=None)
plt.xlabel('체중 5kg 단위', fontsize=16)
plt.ylabel('콜레스테롤', fontsize=16);
plt.title('음주 여부에 따른 체중별 콜레스테롤 양', fontsize=20)
```

그래프를 보면 같은 체중으로 보았을 때 빨간색인 음주 인원 그래프가 전반적으로 비 음주 인원인 초록색 그래프보다 

높은 것을 확인할 수 있습니다.

100kg 이 넘는 경우를 제외하고 100kg 이하에서는 음주를 하는 사람이 콜레스테롤이 더 높은 것을 확인 할 수 있습니다. 


```python
df_nodrink = df_waist[df_waist['음주여부'] == 0]
df_drink = df_waist[df_waist['음주여부'] == 1]
```

```python

plt.figure(figsize=(15,12))
sns.scatterplot(data = df_waist, x = df_waist['체중(5Kg 단위)'],y = df_waist['허리둘레'],hue = '음주여부',palette=['#ACFA58', '#FA5858'],alpha = 0.3)
plt.xlabel('체중', fontsize=16)
plt.ylabel('허리둘레', fontsize=16)
plt.ylim(20,150)
plt.title('음주 여부에 따른 체중별 허리둘레', fontsize=20)
```

위의 산점도 그래프를 보면 음주를 하는 사람은 빨간색, 음주를 하지 않는 사람은 초록색 점인걸 확인할 수 있다.

주로 빨간색점과 초록색점을 비교해보면 체중이 많으면서 허리둘레가 큰 사람들은 빨간색 점인 것을 볼 수 있으며

초록색과 빨간색 점의 분포를 보면 초록색은 미세하지만 왼편 아랫쪽에 분포한 것을 볼 수 있고

빨간점은 초록점보다 오른편 위쪽에 분포한 것을 확인할 수 있다.

```python
plt.figure(figsize=(15,12))
sns.regplot(data = df_nodrink, x = df_nodrink['체중(5Kg 단위)'],y = df_nodrink['총 콜레스테롤'],color = '#ACFA58', scatter = False)
sns.regplot(data = df_drink, x = df_drink['체중(5Kg 단위)'],y = df_drink['총 콜레스테롤'],color = '#FA5858', scatter = False)
plt.legend(
   ('비음주자', '음주자'),
   loc = 'upper right'
)
plt.xlabel('체중 (5kg 단위)', fontsize=16)
plt.ylabel('콜레스테롤양', fontsize=16)
plt.xlim(40,140)
plt.ylim(170,210)
plt.title('음주 여부에 따른 몸무게별 콜레스테롤 양', fontsize=20)
```

위에는 체중별 콜레스테롤 에 대한 선형 회귀 그래프이다.

그래프를 보면 빨간색이 음주자, 초록색이 비음주자이다.

그래프를 봤을때 같은 체중이더라도 음주를 할 경우 더 높은 콜레스테롤 수치를 보일 것 이라고 예측한 것을 확인할 수 있다.

체중이 낮다면 음주 여부에 상관없이 비슷한 콜레스테롤 수치를 보이지만

체중이 많아질수록 두 그룹의 콜레스테롤 수치는 차이가 커지는 것으로 예측하였다.

```python
plt.figure(figsize=(15,12))
sns.regplot(data = df_nodrink, x = df_nodrink['식전혈당(공복혈당)'],y = df_nodrink['총 콜레스테롤'],color = '#ACFA58',scatter = False)
sns.regplot(data = df_drink, x = df_drink['식전혈당(공복혈당)'],y = df_drink['총 콜레스테롤'],color = '#FA5858',scatter = False)
plt.legend(
   ('비음주자', '음주자'),
   loc = 'upper right'
)
plt.xlabel('식전 혈당', fontsize=16)
plt.ylabel('총 콜레스테롤', fontsize=16)
plt.xlim(50,400)
plt.ylim(120,220)
plt.title('음주 여부에 따른 식전혈당, 콜레스테롤 상관분석', fontsize=20)
```

위와 같은 조건으로 X 값을 식전 혈당으로 봤을때도 마찬가지로

같은 혈당을 가진 상태에서 음주 인원이 더 높은 콜레스테롤 수치를 보여주는 것을 확인할 수 있다.
