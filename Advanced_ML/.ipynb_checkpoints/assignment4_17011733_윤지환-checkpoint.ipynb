# 1. 한 노드의 지니 불순도가 보통 그 부모 노드보다 작을까요, 클까요? 일반적으로 작거나 클까요, 항상 작거나 클까요?

일반적으로 작다. 왜냐하면 결정 트리 모델이 학습할 때 훈련 세트를 CART 훈련 알고리즘을 통해서 나누기 때문이다. CART 알고리즘의 비용함수에 의해서 두 서브 셋의 지니 불순도의 합이
가장 낮아지도록 학습하기 때문에 일반적으로 부모 노드보다 작다.

단 항상 작은 것은 아니다. 왜냐하면 노드를 나누었을 때 정확히 반으로 나눈 경우 즉 $ G_i = 1 - \sum_{k=1}^n p_i_k^2 $ 에서 $p_i_k$ = $1/2$ 일 경우에
$G = 1 - (1/2)^2 - (1/2)^2 = 0.5$ 가 나와서 부모의 노드보다 커지는 경우가 있을 수 있다.

예를 들어 나누어진 두개의 노드 중에서 오른쪽 노드의 샘플의 수가 0일 경우 CART 알고리즘의 비용함수 $ J(k, t_k) = \dfrac{m_{\text{left}}}{m}G_\text{left} + \dfrac{m_{\text{right}}}{m}G_{\text{right}}\\$ 에서
오른쪽의 $\dfrac{m_{\text{right}}}{m}$가 0이기 때문에 CART알고리즘의 비용함수 결과는 $\dfrac{m_{\text{left}}}{m}G_\text{left}$ 이다.

이때 $\dfrac{m_{\text{left}}}{m}G_\text{left}$ 이 값이 부모노드의 지니불순도보다 낮을 수 있기 때문에 CART 알고리즘의 비용함수 상으로는 문제가 없다.
하지만 왼쪽의 노드의 지니 불순도 : $G = 0.5$ 이므로 부모 노드보다 큰 경우가 생길 수 있다.





# 2. 결정 트리가 훈련 세트에 과대적합되었다면 max_depth를 어떻게 조정해야 할까요?

모델의 규제를 가하기 위해 max_depth 를 줄여야한다.

depth가 깊어질수록 훈련 세트를 더욱 세세하게 나누기 때문에 과대적합이 발생할 수 있다.

따라서 과대적합을 방지하기 위해서는 max_depth 를 줄여야한다.

아래의 참고자료 1 을 보면 max_depth가 커질수록 영역을 더 세밀하게 나누는 것을 확인할 수 있다.

# 필요한 라이브러리 import

```python
# 파이썬 ≥3.5 필수
import sys
assert sys.version_info >= (3, 5)

# 사이킷런 ≥0.20 필수
import sklearn
assert sklearn.__version__ >= "0.20"

# 공통 모듈 임포트
import numpy as np
import os

# 노트북 실행 결과를 동일하게 유지하기 위해
np.random.seed(42)

# 깔끔한 그래프 출력을 위해
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

# 그림을 저장할 위치
PROJECT_ROOT_DIR = "."
CHAPTER_ID = "training_linear_models"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images", CHAPTER_ID)
os.makedirs(IMAGES_PATH, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("그림 저장:", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)

# 불필요한 경고를 무시합니다 (사이파이 이슈 #5998 참조)
import warnings
warnings.filterwarnings(action="ignore", message="^internal gelsd")
```

# 데이터셋 설정

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=1000, noise=0.4, random_state=42) # noise 에 따라서 데이터의 분포를 선택할 수 있다.

def plot_dataset(X, y, axes):
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bs")
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^")
    plt.axis(axes)
    plt.grid(True, which='both')
    plt.xlabel(r"$x_1$", fontsize=20)
    plt.ylabel(r"$x_2$", fontsize=20, rotation=0)
plot_dataset(X, y, [-1.5, 2.5, -1, 1.5])
plt.show()
```

```python
from matplotlib.colors import ListedColormap

def plot_decision_boundary(clf, X, y, axes=[0, 7.5, 0, 3], iris=True, legend=False, plot_training=True):
    x1s = np.linspace(axes[0], axes[1], 100)
    x2s = np.linspace(axes[2], axes[3], 100)
    x1, x2 = np.meshgrid(x1s, x2s)
    X_new = np.c_[x1.ravel(), x2.ravel()]
    y_pred = clf.predict(X_new).reshape(x1.shape)
    custom_cmap = ListedColormap(['#fafab0','#9898ff','#a0faa0'])
    plt.contourf(x1, x2, y_pred, alpha=0.3, cmap=custom_cmap)
    if not iris:
        custom_cmap2 = ListedColormap(['#7d7d58','#4c4c7f','#507d50'])
        plt.contour(x1, x2, y_pred, cmap=custom_cmap2, alpha=0.8)
    if plot_training:
        plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo", label="Iris setosa")
        plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs", label="Iris versicolor")
        plt.plot(X[:, 0][y==2], X[:, 1][y==2], "g^", label="Iris virginica")
        plt.axis(axes)
    if iris:
        plt.xlabel("Petal length", fontsize=14)
        plt.ylabel("Petal width", fontsize=14)
    else:
        plt.xlabel(r"$x_1$", fontsize=18)
        plt.ylabel(r"$x_2$", fontsize=18, rotation=0)
    if legend:
        plt.legend(loc="lower right", fontsize=14)
```

# train, test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, shuffle= 42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# DecisionTreeClassifier 베이스라인 모델

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf_base = DecisionTreeClassifier(random_state=42)
tree_clf_base.fit(X_train, y_train)
```

```python
tree_clf_base.get_params()
```

```python
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

def model_eval(pred, answer):
    print(f'accuracy_score : {accuracy_score(pred,answer)}')
    print(f'recall_score : {recall_score(pred,answer)}')
    print(f'precision_score : {precision_score(pred,answer)}')
    print(f'f1_score : {f1_score(pred,answer)}')
    print(f'coufusion_matrix : ')
    print(confusion_matrix(pred,answer))
```

```python
model_eval(tree_clf_base.predict(X_test),y_test)
```

# max_depth, min_sample_leaf 에 따른 예측 결과 시각화

```python
tree_clf_depth_2 = DecisionTreeClassifier(max_depth= 2,random_state=42)
tree_clf_depth_4 = DecisionTreeClassifier(max_depth= 4,random_state=42)
tree_clf_depth_6 = DecisionTreeClassifier(max_depth= 6,random_state=42)
tree_clf_depth_8 = DecisionTreeClassifier(max_depth= 8,random_state=42)

tree_clf_depth_2.fit(X_train, y_train)
tree_clf_depth_4.fit(X_train, y_train)
tree_clf_depth_6.fit(X_train, y_train)
tree_clf_depth_8.fit(X_train, y_train)
```

```python
tree_clf_depth_2 = DecisionTreeClassifier(max_depth= 2,random_state=42)
tree_clf_depth_4 = DecisionTreeClassifier(max_depth= 4,random_state=42)
tree_clf_depth_6 = DecisionTreeClassifier(max_depth= 6,random_state=42)
tree_clf_depth_8 = DecisionTreeClassifier(max_depth= 8,random_state=42)

tree_clf_depth_2.fit(X_train, y_train)
tree_clf_depth_4.fit(X_train, y_train)
tree_clf_depth_6.fit(X_train, y_train)
tree_clf_depth_8.fit(X_train, y_train)
```

# 참고자료 1

```python


fig, axes = plt.subplots(ncols=4, figsize=(10, 4), sharey=True)
plt.sca(axes[0])
plot_decision_boundary(tree_clf_depth_2, X_test, y_test, axes=[-1.5, 2.4, -1, 1.5], iris=False)
plt.title("max_depth = 2", fontsize=14)
plt.sca(axes[1])
plot_decision_boundary(tree_clf_depth_4, X_test, y_test, axes=[-1.5, 2.4, -1, 1.5], iris=False)
plt.title("max_depth = 4", fontsize=14)
plt.ylabel("")
plt.sca(axes[2])
plot_decision_boundary(tree_clf_depth_6, X_test, y_test, axes=[-1.5, 2.4, -1, 1.5], iris=False)
plt.title("max_depth = 6", fontsize=14)
plt.ylabel("")
plt.sca(axes[3])
plot_decision_boundary(tree_clf_depth_8, X_test, y_test, axes=[-1.5, 2.4, -1, 1.5], iris=False)
plt.title("max_depth = 8", fontsize=14)
plt.ylabel("")

save_fig("max_depth_plot")
plt.show()
```

```python
tree_clf_leaf_2 = DecisionTreeClassifier(min_samples_leaf= 2,random_state=42)
tree_clf_leaf_4 = DecisionTreeClassifier(min_samples_leaf= 4,random_state=42)
tree_clf_leaf_6 = DecisionTreeClassifier(min_samples_leaf= 6,random_state=42)
tree_clf_leaf_8 = DecisionTreeClassifier(min_samples_leaf= 8,random_state=42)

tree_clf_leaf_2.fit(X_train, y_train)
tree_clf_leaf_4.fit(X_train, y_train)
tree_clf_leaf_6.fit(X_train, y_train)
tree_clf_leaf_8.fit(X_train, y_train)
```

```python
fig, axes = plt.subplots(ncols=4, figsize=(10, 4), sharey=True)
plt.sca(axes[0])
plot_decision_boundary(tree_clf_leaf_2, X_test, y_test, axes=[-1.5, 2.4, -1, 1.5], iris=False)
plt.title("max_leaf = 2", fontsize=14)
plt.sca(axes[1])
plot_decision_boundary(tree_clf_leaf_4, X_test, y_test, axes=[-1.5, 2.4, -1, 1.5], iris=False)
plt.title("max_leaf = 4", fontsize=14)
plt.ylabel("")
plt.sca(axes[2])
plot_decision_boundary(tree_clf_leaf_6, X_test, y_test, axes=[-1.5, 2.4, -1, 1.5], iris=False)
plt.title("max_leaf = 6", fontsize=14)
plt.ylabel("")
plt.sca(axes[3])
plot_decision_boundary(tree_clf_leaf_8, X_test, y_test, axes=[-1.5, 2.4, -1, 1.5], iris=False)
plt.title("max_leaf = 8", fontsize=14)
plt.ylabel("")

save_fig("min_leaf_plot")
plt.show()
```

# GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

# GridSearchCV 를 통해 파라미터 튜닝을 진행
tuned_parameters = {
        'min_impurity_decrease': np.arange(0.0001,0.001,0.0001),
        'max_depth': range(5,20,1),
        'min_samples_split': range(2,100,10)
}


DTC_tuned_1 = GridSearchCV(DecisionTreeClassifier(random_state=42), tuned_parameters, cv=5, scoring="accuracy", verbose=2, refit=True)
DTC_tuned_1.fit(X_train, y_train)
```

```python
print(DTC_tuned_1.cv_results_['mean_test_score'])
```

```python
print(DTC_tuned_1.best_params_)
```

```python
model_eval(DTC_tuned_1.predict(X_test),y_test)
```
